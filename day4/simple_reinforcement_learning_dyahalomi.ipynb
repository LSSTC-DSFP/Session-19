```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=36) 
matplotlib.rc('ytick', labelsize=36) 

# use latex for font rendering
matplotlib.rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
matplotlib.rc('text', usetex=True)
```

```python
def action(prob):
    return np.random.binomial(1, prob)
```

```python
def init(n_init, p1, p2, p3):
    
    success_a1 = 0
    success_a2 = 0
    success_a3 = 0
    tries_array = [0, 0, 0]
    for ii_init in range(0, n_init):
        success_a1 += action(p1)
        success_a2 += action(p2)
        success_a3 += action(p3)
        
        tries_array[0] += 1
        tries_array[1] += 1
        tries_array[2] += 1
        
        
    return([success_a1, success_a2, success_a3], tries_array)
        
```

```python
def greedy(prob_array, success_array_init, tries_array_init, t):

    
    best_prob = np.max(prob_array)
    
    success_array = success_array_init
    tries_array = tries_array_init
    results_array = np.array(success_array)/np.array(tries_array)
    
    #print(success_array)
    #print(tries_array)
    #print(results_array)
    #print('')
    regret_array = []
    regret_sum = 0
    for ii in range(0, t):
        best_init_index = np.argmax(results_array)
        
        regret = best_prob - prob_array[best_init_index]
        regret_sum += regret
        regret_array.append(regret_sum)
        
        #print('best index = ' + str(best_init_index))
        
        #in greedy we always chose the best action...
        action_result = action(prob_array[best_init_index])
        #print('action result = ' + str(action_result))
        
        success_array[best_init_index] += action_result
        tries_array[best_init_index] += 1
        
        results_array = np.array(success_array)/np.array(tries_array)
        
        
        
        #print(success_array)
        #print(tries_array)
        #print(results_array)
        #print('')
        
    return success_array, tries_array, results_array, regret_array
        
        
    
    
    
    
    
```

```python
n_inits = [1, 2, 3]
ts = [500, 1000, 5000]
epochs = 20


regret_array = []
nsteps = []
for epoch in range(0, epochs):
    for ii in range(0, len(n_inits)):
        ii_init = n_inits[ii]
        for t in ts:    
            init_out = init(ii_init, 0.95, 0.9, 0.1)
            greedy_out = greedy([0.95, 0.9, 0.1], init_out[0], init_out[1], t)

            regret_array.append(greedy_out[3])
            nsteps.append(t)

        
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize = [27,13], sharey=True)

for ii in range(0, len(regret_array)):
    regret = regret_array[ii]
    step_num = np.arange(0, nsteps[ii])
    if nsteps[ii] == ts[0]:
        ax[0].plot(step_num, np.log(np.cumsum(regret)))
    elif nsteps[ii] == ts[1]:
        ax[1].plot(step_num, np.log(np.cumsum(regret)))
    elif nsteps[ii] == ts[2]:
        ax[2].plot(step_num, np.log(np.cumsum(regret)))
    

ax[0].set_xlabel('step number', fontsize=27)
ax[1].set_xlabel('step number', fontsize=27)
ax[2].set_xlabel('step number', fontsize=27)
ax[0].set_ylabel('ln(cumsum(regret))', fontsize=27)

ax[0].grid(True)  # Turn on the grid for the first subplot
ax[1].grid(True)  # Turn on the grid for the 2nd subplot
ax[2].grid(True)  # Turn on the grid for the 3rd subplot


fig.suptitle('greedy algorithm', fontsize=54)
fig.tight_layout()
```

```python
def egreedy(prob_array, success_array_init, tries_array_init, t, e):

    
    best_prob = np.max(prob_array)
    
    success_array = success_array_init
    tries_array = tries_array_init
    results_array = np.array(success_array)/np.array(tries_array)
    
    #print(success_array)
    #print(tries_array)
    #print(results_array)
    #print('')
    regret_array = []
    regret_sum = 0
    for ii in range(0, t):
        best_init_index = np.argmax(results_array)
        
        regret = best_prob - prob_array[best_init_index]
        regret_sum += regret
        regret_array.append(regret_sum)
        
        #print('best index = ' + str(best_init_index))
        
        #in egreedy we chose the best action with prob 1-e...
        random_val = np.random.random()
        if random_val < 1-e:
            action_result = action(prob_array[best_init_index])
        else: #pick a random index
            random_index = np.random.randint(0, 3)
            action_result = action(prob_array[random_index])

            
        success_array[best_init_index] += action_result
        tries_array[best_init_index] += 1
        
        results_array = np.array(success_array)/np.array(tries_array)
        
        
        
        #print(success_array)
        #print(tries_array)
        #print(results_array)
        #print('')
        
    return success_array, tries_array, results_array, regret_array
        
        
    
    
    
    
    
```

```python
n_inits = [1, 2, 3]
ts = [500, 1000, 5000]
epochs = 20


regret_array = []
nsteps = []
for epoch in range(0, epochs):
    for ii in range(0, len(n_inits)):
        ii_init = n_inits[ii]
        for t in ts:    
            init_out = init(ii_init, 0.95, 0.9, 0.1)
            egreedy_out = egreedy([0.95, 0.9, 0.1], init_out[0], init_out[1], t, 0.05)

            regret_array.append(egreedy_out[3])
            nsteps.append(t)

        
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize = [27,13], sharey=True)

for ii in range(0, len(regret_array)):
    regret = regret_array[ii]
    step_num = np.arange(0, nsteps[ii])
    if nsteps[ii] == ts[0]:
        ax[0].plot(step_num, np.log(np.cumsum(regret)))
    elif nsteps[ii] == ts[1]:
        ax[1].plot(step_num, np.log(np.cumsum(regret)))
    elif nsteps[ii] == ts[2]:
        ax[2].plot(step_num, np.log(np.cumsum(regret)))
    

ax[0].set_xlabel('step number', fontsize=27)
ax[1].set_xlabel('step number', fontsize=27)
ax[2].set_xlabel('step number', fontsize=27)
ax[0].set_ylabel('ln(cumsum(regret))', fontsize=27)

ax[0].grid(True)  # Turn on the grid for the first subplot
ax[1].grid(True)  # Turn on the grid for the 2nd subplot
ax[2].grid(True)  # Turn on the grid for the 3rd subplot


fig.suptitle('e-greedy algorithm', fontsize=54)
fig.tight_layout()
```
